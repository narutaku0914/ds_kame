第13回 フィルタ操作はDataFrameの操作で最も頻出かつ重要な操作の一つ

```python
import pandas as pd
df = pd.read_csv('train.csv')
df
```

```python
# 生存者のレコードだけに絞りたい
filter = df['Survived'] == 1  # SQLでいうwhere
df[filter]
```

元のDataFrameは上書きされない！

```python
df[df['Survived'] == 1] # 普通は代入せずこの書き方
```

```python
df[df['Survived'] == 1].describe()
```

```python
df.describe()
```

```python
# & や | で複数条件のフィルタリング
df[(df['Age'] >= 60) & (df['Sex'] == 'female')]
```

```python
df[(df['Pclass'] == 1) | (df['Age'] < 10)]
```

```python
# ~(スクィグル)をつけるとNOT演算でフィルタリング
df[~(df['Sex'] == 'female')]
```

値がbooleanのカラムでのフィルタリングで使う

```python
data = [{'Name':'John', 'Survived':True},
        {'Name':'Emily', 'Survived':False},
        {'Name':'Ben', 'Survived':True}]
df = pd.DataFrame(data)
df
```

```python
df[df['Survived']==True]
```

```python
df[df['Survived']] # SurvivedカラムがBooleanなので　== True いらない
```

```python
df[~(df['Survived'])]
```

```python
# indexの変更
#.reset_index()で再度割り振り
df = pd.read_csv('train.csv')
df = df[df['Sex'] == 'male']
df
```

```python
df.reset_index()
```
