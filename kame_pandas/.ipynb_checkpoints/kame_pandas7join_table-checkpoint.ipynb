第16回 複数の表形式のDataFrameを結合 → こちらもデータサイエンス必須操作!

```python
import pandas as pd
df1 = pd.DataFrame({
    'Key': ['k0', 'k1', 'k2'],
    'A': ['a0', 'a1', 'a2'],
    'B': ['b0', 'b1', 'b2']
})
df2 = pd.DataFrame({
    'Key': ['k0', 'k1', 'k2'],
    'C': ['c0', 'c1', 'c2'],
    'D': ['d0', 'd1', 'd2']
})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)  # merge(右にくっつけたいtable)
```

```python
# 縦にくっつける
pd.concat([df1, df2], axis=0) # デフォルトはaxis=0 → 縦
```

```python
# 横にくっつける
pd.concat([df1, df2], axis=1)
```

concatenateはkey関係なくただガッチャンコ　　　　　　　　　　　　　　　　　　　　　　　　　
縦につなげる方が圧倒的に多い

.merge(                                                                           
    how='inner' : どう結合するか→{‘left’, ‘right’, ‘outer’, ‘inner’}, デフォルトは ‘inner’                                                                          
    on=None : keyにするカラムを指定（どちらのDataFrameにも存在するカラム）指定をしないと共通のカラム  
    left_on=None : leftのDataFrameのkeyにするカラム                                   
    right_on=None                                                                 
    left_index=False : leftのKeyをindexにする場合Trueを指定                           
    right_index=False                                                             
    .                                                                             
    .                                                                             
    .                                                                             
    ) 上の６項目は重要！ 
"""

rightは一応用意されているが使わないと考えるとシンプル → leftに持ってればよい 

```python
# how どう結合するか
df1 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k2'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']})

df2 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k3'],
                    'C': ['c0', 'c1', 'c3'],
                    'D': ['d0', 'd1', 'd3']})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, how='left') # k2ないため　→ 左に合わせ込む
```

```python
# outer どっちもレコード失わず、対応できないとこはNaN, orと同じ
df1.merge(df2, how='outer')
```

```python
# inner 対応できるとこのみで結合, andと同じ
df1.merge(df2, how='inner')
```

実際業務で結合する場合の多くが「大きな表に，小さな表を結合する」ケース                             
ex) タイタニックのデータに対して，追加のデータとしてIDと他の情報(例えば英語を喋るか喋らないかのフラグなど，なんでもいいですが)を付け加えるケース                                                    
→ 「ベースの表を左において，追加の表を右から左にmergeするイメージ」が重要!

よく使うのは’left’と’inner’

```python
# on どう結合するか, なくても共通のカラムがkeyとなるが、指定したほうわかりやすいし安全
df1 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k2'],
                    'ID': ['aa', 'bb', 'cc'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']})

df2 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k3'],
                    'ID': ['aa', 'bb', 'cc'],
                    'C': ['c0', 'c1', 'c3'],
                    'D': ['d0', 'd1', 'd3']})
```

```python
df1
```

```python
df2
```

```python
# keyをkeyにして
df1.merge(df2, on='Key')
```

```python
# IDをkey
df1.merge(df2, on='ID')
```

引数onはなるべく明記して！　他の人がコードを読むときにわかりやすい！

```python
# suffixes(_x, _y)を指定
df1.merge(df2, on='ID', suffixes=('_left', '_right'))
```

```python
# keyにしたいカラムが異なるとき
df1 = pd.DataFrame({ 'Key1': ['k0', 'k1', 'k2'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']})

df2 = pd.DataFrame({ 'Key2': ['k0', 'k1', 'k3'],
                    'C': ['c0', 'c1', 'c3'],
                    'D': ['d0', 'd1', 'd3']})
df1.merge(df2, left_on='Key1', right_on='Key2')
```

★カラムが増えると管理できなくなるので，できるだけ同じ名前で管理したほうが無難

```python
# indexをkeyにしたい場合
df1.merge(df2, left_index = True, right_index = True)
```

分かりづらいしIndexは変わりやすい

```python
# joinはmergeと同じだから無視してOK
df1 = pd.DataFrame({ 'Key1': ['k0', 'k1', 'k2'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']})

df2 = pd.DataFrame({ 'Key2': ['k0', 'k1', 'k3'],
                    'C': ['c0', 'c1', 'c3'],
                    'D': ['d0', 'd1', 'd3']})
```

```python
df1.join(df2)
```

```python
df1.merge(df2, left_index = True, right_index = True)
```
