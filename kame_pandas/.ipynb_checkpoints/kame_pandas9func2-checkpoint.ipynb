第18回 その他頻出関数(to_csv, iterrows, sort_values)

```python
import pandas as pd
df = pd.read_csv('train.csv')
```

```python
# .to_csv() DataFrameをcsvで保存
df['adult'] = df['Age'].apply(lambda x: x > 20)
df
```

```python
df.to_csv('train_w_adult.csv')
```

```python
df = pd.read_csv('train_w_adult.csv')
df
```

index情報が"Unnamed:0"に保存　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　
これを繰り返すと、上記の絡むがどんどん追加される

```python
# 基本常にindex=Falseを指定してcsv形式に保存
df = pd.read_csv('train.csv')
df['Adult'] = df['Age'].apply(lambda x: x > 20)
df.to_csv('train_w_adult.csv', index=False)
```

```python
df = pd.read_csv('train_w_adult.csv')
df.head(3)
```

型によっては文字列で保存されてしまうことがあるので注意！

```python
df = pd.DataFrame({'A': [['a', 'b'], 2, 3], 'B': [['c', 'd'], 5, 6]})
df
```

```python
# .iterrows()　DataFrameをforでiteration
df = pd.read_csv('train.csv')
for idx, row in df.iterrows():  # dictの.items()と同じ挙動
    if row['Age'] > 40 and row['Pclass'] ==3 and row['Sex'] == 'male' and row['Survived'] == 1:
        print('{} is very lucky guy!'.format(row['Name']))
```

使い分け;　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　
apply → 各レコードの処理の結果を別カラムに保存　　　　　　　　　　　　　　　　　　　　　　　　
iterrows → 処理をしたいだけ(値を返さない)

```python
# .sort_values() 特定のカラムでsort
df.sort_values('Age')
```

ファイルを保存できるようになると，もうほとんどのExcel処理をPythonでできるようになっているはず！
