第24回 Seaborn -　histgram, scatter編

おしゃれなグラフ → DStistはこれがデファクトスタンダード

```python
import seaborn as sns
# 以下は変わらず
%matplotlib inline
```

```python
# sns.plot() 頻出！
import pandas as pd
df = pd.read_csv('train.csv')
df = df.dropna(subset=['Age'])
sns.distplot(df['Age'])
```

```python
sns.set()
sns.distplot(df['Age'])
```

```python
# sns.jointplot() 2変数の分布
sns.jointplot(x='Age', y='Fare', data=df)
```

```python
sns.jointplot(x='Age', y='Fare', data=df, kind='hex') # kindのdefaultはscatter
```

```python
# sns.pairplot()で複数のカラムの分布を一発で表示 超重要
sns.pairplot(df)
```

めちゃくちゃ使う

```python
sns.pairplot(df, hue='Survived', kind='scatter', plot_kws={'alpha':0.5})
```

ブログ通りに書いたが動かないのは不明。エラー文をようチェック

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', plot_kws={'alpha':0.5})
```

普通は必要なカラムだけ抽出
