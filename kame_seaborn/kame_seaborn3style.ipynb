```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')

sns.set(context='poster') # 文字が大きくなる
df = df.dropna(subset=['Age'])
sns.distplot(df['Age'])
```

```python
sns.set(context='notebook')
sns.distplot(df['Age'])
```

```python
sns.set(context='talk')
sns.distplot(df['Age'])
```

```python
sns.set(context='paper')　# set(), set_xxx()はnotebook全体に影響
sns.distplot(df['Age'])
```

paper < notebook < talk < poster で大きくなる　基本はnotebook(default)でOK

poster用，プレゼン用にグラフを描画する際はこのcontext引数を指定して実行すると，それ用に最適化されたグラフが描画されるのでめちゃくちゃ便利！

```python
# style引数でグラフ全体のstyleを指定
# ‘darkgrid(default)’, ‘whitegrid’, ‘dark’, ‘white’, ‘ticks’
sns.set_style('whitegrid')
sns.distplot(df['Age'])
```

```python
# palette 色を指定
sns.set(palette='bright')
sns.violinplot(x='Pclass', y='Age', data=df)
```

```python
# despine() 軸や枠を落とす
sns.set(palette='bright')
sns.set_style('whitegrid')
sns.violinplot(x='Pclass', y='Age', data=df)
sns.despine() #
```

```python
sns.set(palette='bright')
sns.set_style('whitegrid')
sns.violinplot(x='Pclass', y='Age', data=df)
sns.despine(top = False, right = False) # kwdはすべて逆がsdefault
```

context, style, paletteは　set_context()でもOK
