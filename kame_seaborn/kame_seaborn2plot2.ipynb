第25回 barplot, boxplot, swarmplot etc...

```python
import pandas as pd
import seaborn as sns
%matplotlib inline

df = pd.read_csv('train.csv')
df.head()
```

```python
# barplot() 棒グラフ
# Survived指定で、Ageの平均を棒グラフで比較
sns.barplot(x='Survived', y='Age', data=df)
```

Seabornではデフォルトで95%信頼区間(Confidence Intervals; CI)がError Bar
CIを変えたい、標準偏差を使いたい場合はci引数

```python
# 平均値以外の値を比較したい場合estimator引数
import numpy as np
sns.barplot(x='Survived', y='Age', data=df, estimator=np.median)
```

numpyの関数で行ける

```python
# countplot() データ数を比較する
sns.countplot(x='Sex', data=df, hue='Survived')
```

```python
# boxplot() カテゴリ別に比較 箱ひげ
sns.boxplot(x='Pclass', y='Age', data=df)
```

```python
sns.boxplot(x='Pclass', y='Age', data=df, hue='Survived')
```

Seabornでグラフを作ってると簡単に生成できるがゆえに複雑になってしまうので注意!

```python
# violinplot() データの分布を可視化
sns.violinplot(x='Pclass', y='Age', data=df)
```

情報量が増えるので，若干ビジーなのと見たことない人からすると見方がわからないので注意

```python
sns.violinplot(x='Pclass', y='Age', data=df, hue='Survived')
```

violinplotを見ると乗客数の差がそれぞれのPclassの幅の差に反映されてない　　　　　　　　　　　
→ violinplotで表示しているのはあくまでも密度であり実際の数の大きさではないから　

```python
# swarmplot() 本当の分布が確認できる
sns.set()
sns.swarmplot(x='Pclass', y='Age', data=df)
```

統計量はでてこないのでboxplotやviolinplotと合わせて使うといい

```python
sns.swarmplot(x='Pclass', y='Age', data=df, size=4) # markersizeを小さくした　defaultは5
```

```python
sns.swarmplot(x='Pclass', y='Age', data=df, size=4, hue='Survived')
```

boxplotとswarmplotを組み合わせると非常にGood　　　　　　　　　　　　　　　　　　　　　　
★「そのグラフで何を伝えたいのか」が一番重要で，それを伝えるために最低限必要な情報のみグラフに入れるようにする
