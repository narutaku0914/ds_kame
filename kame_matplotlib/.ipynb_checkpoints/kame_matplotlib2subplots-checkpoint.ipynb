第21回　matplotlibのsubplotとオブジェクト指向を使って複数のグラフを並べる

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# subplot
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = x*x

plt.subplot(1, 2, 1)  # row, column, index
plt.plot(x, y1)

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.show()
```

```python
plt.subplot(2, 1, 1)  # row, column, index
plt.plot(x, y1)

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.show()
```

```python
# object指向
fig = plt.figure()
type(fig)
```

```python
fig = plt.figure() # plt.figureクラスのインスタンス生成
ax1 = fig.add_subplot(1, 2, 1) # axオブジェクト生成
ax2 = fig.add_subplot(1, 2, 2) ß

ax1.plot(x, y1)
ax2.plot(x, y2)
plt.show()
```

```python
# 1行2列のplot生成 axesにはそれぞれのオブジェクトがリストで返される
fig, axes = plt.subplots(nrows=1, ncols=2)
axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y1)
axes[1].plot(x, y2)
```

よくこのaxesをfor文で回して大量のグラフや画像を描画すること有　その時にめちゃくちゃ便利

＊ subplots()のnrows >= 2のとき == 2行以上のsubplotsのときaxesは多次元array

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
axes[1, 2].plot(x, y2)
```

```python
# set_xxで描画情報追加
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y1, label='something')
axes[1].plot(x, y1)
axes[0].set_xlabel('xlabel1')
axes[0].set_ylabel('ylabel1')
axes[0].set_title('plot title')
axes[0].set_xticks([-3, -2, -1, 3])
axes[0].set_yticks([0, 10, 20])
axes[0].legend()
axes[1].axis('off')
```

★ fig, axes = plt.subplots(nrows= , ncols= )　が最重要！
