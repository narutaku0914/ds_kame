第22回 PNG, PDFで保存

必要なコードは１~4行

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x)*2
fig, axes = plt.subplots()
axes.plot(x, y1)
plt.show()
```

```python
# figsize引数とdpi引数を使ってグラフのサイズを変える
fig, axes = plt.subplots(figsize=(1, 1), dpi=100) # figsizeは(width, height), dots per inch
axes.plot(x, y1)
```

```python
fig, axes = plt.subplots(figsize=(10, 3))
axes.plot(x, y1)
```

```python
# pngで保存 一番手軽で一番使う
fig, axes = plt.subplots(2, 1, figsize=(10,3))
axes[0].plot(x, y1)
axes[1].plot(x, x*x)

axes[0].set_title('first')
axes[1].set_title('second')

axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')

fig.savefig('savefig_sample.png')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10,3))
axes[0].plot(x, y1)
axes[1].plot(x, x*x)

axes[0].set_title('first')
axes[1].set_title('second')

axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')

plt.tight_layout()  # 重ならないよう調整

fig.savefig('savefig_sample.png')
```

```python
# pdf保存
from matplotlib.backends.backend_pdf import PdfPages

pdf = PdfPages('saving_sample.pdf')
```

```python
pdf.savefig(fig)

pdf.close()
```

pdfで保存する際の一番の利点は「大量のグラフを一つのファイルで保存」できること

```python
pdf = PdfPages('savemultifig_sample.pdf')
for i in range(1, 10):
    fig, axes = plt.subplots()
    axes.plot(x, y1 + x*i)
    axes.set_title('ID:#{}'.format(i))
    
    pdf.savefig(fig)
    
pdf.close() # pdfのときこれ大事！
```

pngは必修, pdfは余力
