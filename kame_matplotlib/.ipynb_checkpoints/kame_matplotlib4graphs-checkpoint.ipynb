第23回 色々なグラフを描画

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-3, 3, 10)
y = np.exp(x)
plt.plot(x, y)
```

```python
plt.plot(x, y, color='red', lw=5, ls='--', marker='o', markersize=15, markerfacecolor='yellow', markeredgecolor='blue', markeredgewidth=4, alpha=0.5)
plt.show()
```

```python
"""
color : グラフの線の色　⇨　’red’や’green’のように色の名前を指定 ’r’や’g’のように頭文字でもOK
lw (line width) : 線の太さ　⇨　数字
ls (line style) : 線の種類　⇨　’-‘や’–‘のように指定 よく使うのはこの二つのどちらか
marker : マーカーの種類　⇨　’o’や’x’ように指定
markersize : マーカーの大きさ　⇨　数字
markerfacecolor : マーカーの色　⇨　colorと同じ
markeredgecolor : マーカーの枠に色　
markeredgewidth : マーカーの枠の太さ
alpha : plotの透明度　⇨　0から1の間をfloatで指定します．0に近い程透明度が上がります．
"""
```

DSは散布図，ヒストグラム，箱ひげ図などのグラフであることが多い → その中でも散布図が一番！

```python
# plt.scatter() 散布図
import pandas as pd
df = pd.read_csv('train.csv')
plt.scatter(df['Age'], df['Fare'], alpha=0.3) # 重なりをわかりやすくするための工夫
plt.show()
```

```python
# plt.hist() ヒストグラム
plt.hist(df['Age'])
plt.show()
```

```python
plt.hist(df['Age'], bins=50) # bin: 階級
plt.show()
```

```python
# plt.boxplot() 箱ひげ図
df = df.dropna(subset=['Age']) # NaNがあると表示できないのでdrop
plt.boxplot(df['Age'])
plt.show()
```
