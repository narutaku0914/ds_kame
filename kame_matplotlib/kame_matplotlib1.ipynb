```python
import numpy as np
import matplotlib.pyplot as plt   # 基本はこの読み込み
```

```python
x = np.linspace(-3, 3, 10)
y = np.exp(x)
print(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r')
```

```python
plt.plot(x, y, 'g^')
```

```python
plt.plot(x, y, '--')
```

```python
plt.plot(x, y, 'bo')
```

```python
plt.plot(x, y)
# ラベル
plt.xlabel('Effort')
plt.ylabel('Earning')
plt.title('This is how your effort earns')
```

plt.show()を追加すると最後に描画処理が発生するのでこのような無駄な返り値がprintされない

```python
# label= でplotにlabelをつけ、.legend()で凡例つける
plt.plot(x, y, label='Earning with effort')
plt.legend()
# ticksで任意の目盛りつけ
plt.xticks(np.arange(-3, 4, 0.5))
plt.yticks(np.arange(0, 25, 5))
plt.show()
```

```python
# 複数plotするには単にもう一つ plt.plot() を実行するだけ
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x)*2
plt.plot(x, y1, label='first')
plt.plot(x, y2, label='second')
plt.axis('off')
plt.show()
```
