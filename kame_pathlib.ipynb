第31回 os, pathlibモジュールでPath操作

前回のようなパス指定(文字列操作)OSによってはうまく動かない        
文字列ではなくてosモジュールを使うことで，安全にpathを操作可能

```python
import os
from pathlib import Path
```

```python
# pathlibを使ってフォルダおよびファイル構造を取得
p = Path('gaussian_filtered_images') #Path instance
list(p.iterdir())
```

glob関数と * (アスタリスク)を使うことで特定のパターンにマッチするファイルパスのリストを取得可能　　
→ Pathクラスに glob() 関数が装備!

```python
# glob関数を使ってpngファイルのpathリストを取得
list(p.glob('*/*.png')) # pである程度よみこんでいるためこれだけでOK
```

```python
# ファイルパスを文字列で取得するにはPathオブジェクトの .as_posix() 関数をコール
list(p.glob('*/*.png'))[0].as_posix()
```

```python
# osモジュールでpathをjoin・split
p = Path('gaussian_filtered_images')

folder = list(p.iterdir())[3] # 任意のフォルダ選択
print(folder)
print(type(folder))
```

```python
split_list = os.path.split(folder)
print(split_list)
file_path = os.path.join(folder, 'test.txt')
print(file_path)
```

```python
# os.path.exits()でフォルダの有無を確認して無ければ os.makedirs()でフォルダを作成する
p = Path('gaussian_filtered_images')
new_folder_name = 'new_folder'
new_folder_path = os.path.join(p, new_folder_name)

print(new_folder_path)
```

```python
os.path.exists(new_folder_path)
```

```python
if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)
    
os.path.exists(new_folder_path)
```

```python
os.makedirs(new_folder_path)
```

os.path.exits()でフォルダの有無を確認して無ければ os.makedirs()でフォルダを作成　はめっちゃ大事な流れ

データサイエンスは当然データを扱うのでファイルパスを操作する機会が非常に多い　　　
→ この操作大事！
